### - HIPÓTESIS O CUESTIONES DE TRABAJO QUE AFECTAN A UNA SOLA VARIABLE

_**1. ¿Qué tipo de establecimiento hotelero es el más común en la Comunidad Foral de Navarra?**_

Calculamos la frecuencia de cada tipo de establecimiento hotelero en función de la MODALIDAD:

- Primeramente de forma numérica:

```python
# De forma numérica:
hoteles_Navarra['MODALIDAD'].value_counts()
```

- Ahora de forma visual. Lo podemos hacer calculando la frecuencia en valores absolutos:

```python
# Vamos a usar un gráfico de barras en el que la altura de las barras represente la frecuencia absoluta de cada 
# uno de los tipos de establecimientos hoteleros (variable categórica nominal MODALIDAD).
```

```python
# Se genera el gráfico inicial y se almacena una referencia al mismo en 'ax'
ax = hoteles_Navarra['MODALIDAD'].value_counts().plot.bar(figsize=(10, 5), color='green')

# Se emplea la referencia para hacer modificaciones en el gráfico
ax.set_title('Frecuencia de cada tipo de establecimineto hotelero', fontsize=20)
ax.set_xlabel('Modalidad de establecimiento', fontsize=15)
ax.set_ylabel('Frecuencia absoluta', fontsize=15)
```

Como se puede ver, la mayoría de establecimientos hoteleros de la Comunidad Foral de Navarra (casi 140) tienen la modalidad de Hostal, seguidos por Hoteles y Pensiónes, con cifras muy similares (92 y 86 respectivamente). A continuación estarían los Hoteles rurales (con 30 establecimientos). Por último tenemos que apenas hay Hoteles-apartamento (solamente 2).

- O calculando la frecuencia relativa de cada modalidad de establecimiento hotelero, sin más que dividir las frecuencias absolutas por el total de establecimientos hoteleros:

```python
# Vamos a usar como antes un gráfico de barras, pero en el que ahora la altura de las barras represente la 
# frecuencia relativa de cada uno de los tipos de establecimientos hoteleros (variable categórica nominal 
# MODALIDAD).
```

```python
# Se genera el gráfico inicial y se almacena una referencia al mismo en 'ax'
ax = (hoteles_Navarra['MODALIDAD'].value_counts()/len(hoteles_Navarra)).plot.bar(figsize=(10, 5), color='green')

# Se emplea la referencia para hacer modificaciones en el gráfico
ax.set_title('Frecuencia de cada tipo de establecimineto hotelero', fontsize=20)
ax.set_xlabel('Modalidad de establecimiento', fontsize=15)
ax.set_ylabel('Frecuencia relativa', fontsize=15)
```

La gráfica es la misma que la anterior pero ahora los resultados de la frecuencia son relativos, es decir, se ha dividido la frecuencia de cada tipo de establecimiento entre el total de establecimientos. Se obtiene que los Hostales representan casi el 40%, los Hoteles algo más del 25%, seguidos por las Pensiones con poco menos del 25%, los Hoteles rurales no llegan al 10%, y por último estarían los Hoteles-apartamento con menos de un 1%.

_**2. ¿Cuál es la distribución de frecuencia de los establecimientos hoteleros de la Comunidad Foral de Navarra en función del número de plazas del que disponen?**_

Como la variable PLAZAS es cuantitativa y por lo tanto tiene un orden natural, vamos a utilizar una representación en histograma para visualizar la distribución de frecuencias que nos piden:

```python
# Vamos a usar un histograma en el que el área de las barras represente la frecuencia absoluta de 
# establecimientos hoteleros en función del número de plazas (variable cuantitativa discreta PLAZAS).
```

```python
# Se genera el gráfico inicial y se almacena una referencia al mismo en 'ax'
ax = hoteles_Navarra['PLAZAS'].plot.hist(figsize=(10, 7))

# Se emplea la referencia para hacer modificaciones en el gráfico
ax.set_title('Frecuencia de los establecimientos hoteleros en función del número de plazas', fontsize=20)
ax.set_xlabel('Número de plazas del establecimiento hotelero', fontsize=15)
ax.set_ylabel('Frecuencia absoluta', fontsize=15)
```

Se puede observar que la gran mayoría de los establecimientos hoteleros de la Comunidad Foral de Navarra tienen menos de 50 plazas.

Vamos a quedarnos con los establecimientos que tienen menos de 50 plazas, ya que son la gran mayoría, para ver su distribución de frecuencia con más claridad:

```python
# Ahora vamos a usar un histograma en el que el área de las barras represente la frecuencia absoluta de 
# establecimientos hoteleros de menos de 50 plazas en función del número de plazas (variable cuantitativa 
# discreta PLAZAS).
```

```python
# Se genera el gráfico inicial y se almacena una referencia al mismo en 'ax'
ax = hoteles_Navarra[hoteles_Navarra['PLAZAS'] < 50]['PLAZAS'].plot.hist(figsize=(10, 7))

# Se emplea la referencia para hacer modificaciones en el gráfico
ax.set_title('Frecuencia de los establecimientos hoteleros en función del número de plazas', fontsize=20)
ax.set_xlabel('Número de plazas del establecimiento hotelero', fontsize=15)
ax.set_ylabel('Frecuencia absoluta', fontsize=15)
```

Como se puede comprobar, el rango de plazas en el que el número de establecimientos hoteleros en la Comunidad Foral de Navarra es mayor, está entre 5 y 15 plazas. A partir de ese valor, el número de establecimientos disminuye progresivamente según aumenta el número de plazas.

Podemos representar lo anterior mediante frecuencias relativas, y añadiendo una función de densidad, de la siguiente forma:

```python
sns.distplot(hoteles_Navarra['PLAZAS'])
```

**NOTA:** Habría que tener cuidado con la representación de la función de densidad calculada de esta forma, ya que da la impresión de que hay establecimientos con un número de plazas negativo, lo cual no es real y es imposible.

_**3. Vamos a comprobar la siguiente hipótesis: la mayoría de los establecimientos hoteleros de la Comunidad Foral de Navarra se encuentran en Pamplona, por ser la capital, y en la zona de los Pirineos, por ser la zona más turística junto con Pamplona.**_

Vamos a comenzar calculando la frecuencia de establecimientos hoteleros en función de la ZONA o región:

```python
# Vamos a usar un gráfico de barras en el que la altura de las barras represente la frecuencia absoluta de 
# establecimientos hoteleros en función de la zona o región de la Comunidad Foral de Navarra (variable 
# categórica nominal ZONA).
```

```python
# Se genera el gráfico inicial y se almacena una referencia al mismo en 'ax'
ax = hoteles_Navarra['ZONA'].value_counts().plot.bar(figsize=(10, 7), color='orange')

# Se emplea la referencia para hacer modificaciones en el gráfico
ax.set_title('Frecuencia de los estableciminetos hoteleros en función de la región', fontsize=20)
ax.set_xlabel('Zona o región', fontsize=15)
ax.set_ylabel('Frecuencia absoluta', fontsize=15)
```

Ya sólo con ese gráfico se puede ver que la zona con mayor número de establecimientos hoteleros es la de los Pirineos. En segundo lugar vemos que está la Cuenca de Pamplona.

Veamos qué porcentaje de los establecimientos hoteleros de la Cuenca de Pamplona están en Pamplona capital:

```python
# Usamos ahora un gráfico de barras en el que la altura de las barras represente el tanto por ciento de 
# establecimientos hoteleros sobre el total de los de la Cuenca de Pamplona en función de la localidad a la que 
# pertenecen (variable categórica nominal LOCALIDAD).
```

```python
# Se genera el gráfico inicial y se almacena una referencia al mismo en 'ax'
ax = (
    hoteles_Navarra[hoteles_Navarra['ZONA'] == 'Cuenca de Pamplona']['LOCALIDAD']
    .value_counts()*100/len(
        hoteles_Navarra[hoteles_Navarra['ZONA'] == 'Cuenca de Pamplona']
    )).plot.bar(figsize=(20, 10), color='red')

# Se emplea la referencia para hacer modificaciones en el gráfico
ax.set_title('Distribución por localidad de los estableciminetos hoteleros de la Cuenca de Pamplona', fontsize=20)
ax.set_xlabel('Localidad', fontsize=15)
ax.set_ylabel('Frecuencia relativa (%)', fontsize=15)
```

Podemos observar que cerca del 70% de los estableciemientos hoteleros de la zona de la Cuenca de Pamplona se encuentran en Pamplona capital.

Por lo tanto, a la vista de los dos gráficos anteriores, podemos concluir que se cumple la hipótesis planteada de que la mayoría de los establecimientos hoteleros de la Comunidad Foral de Navarra se encuentran en Pamplona, por ser la capital, y en la zona de los Pirineos, por ser la zona más turística junto con Pamplona.

### - HIPÓTESIS O CUESTIONES DE TRABAJO QUE AFECTAN A LA RELACIÓN DE DOS O MÁS VARIABLES, O A LA RELACIÓN DE UNA VARIABLE Y UNA AGRUPACIÓN SOBRE LA MISMA

_**1. Queremos ver si hay relación entre el número de plazas del establecimiento hotelero y la modalidad de establecimiento de la que se trate, ya que tenemos la hipótesis de que los establecimientos con mayor número de plazas son Hoteles. Además agruparemos dentro de cada modalidad por categoría.**_

Comenzamos calculando la distribución de plazas en función de la modalidad de establecimiento hotelero del que se trate:

```python
# Vamos a usar un gráfico de cajas, con varias cajas (una por cada modalidad de establecimiento hotelero), las 
# cuales nos mostrarán la distribución de la variable plazas para cada modalidad (variable categórica nominal 
# MODALIDAD, variable cuantitativa discreta PLAZAS).
```

```python
sns.set(rc = {'figure.figsize':(15,8)})  # Fijamos el tamaño del gráfico.
sns.boxplot(
    x='MODALIDAD',
    y='PLAZAS',
    data=hoteles_Navarra
).set_title('DISTRIBUCIÓN DE PLAZAS DE LOS ESTABLECIMIENTOS HOTELEROS EN FUNCIÓN DE LA MODALIDAD DE LOS MISMOS')
```

A la vista de los resultados se puede comprobar como se confirma la hipótesis que manejábamos, ya que los establecimientos con mayor número de plazas son Hoteles, los cuales tienen una gran variabilidad en el número de plazas que pueden tener.  
El resto de modalidades de establecimientos tienen menos de 50 ó 60 plazas, excepto algún caso aislado de Hostal.

Si ahora calculamos la distribución de plazas en función de la modalidad de establecimiento hotelero agrupado por categoría:

```python
# Usamos en este caso un gráfico de cajas, con varias cajas (una por cada modalidad y categoría de establecimiento 
# hotelero), las cuales nos mostrarán la distribución de la variable plazas para cada modalidad y categoría 
# (variable categórica nominal MODALIDAD, variable categórica ordinal CATEGORIA, variable cuantitativa discreta 
# PLAZAS).
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.boxplot(
    x='MODALIDAD',
    y='PLAZAS',
    hue='CATEGORIA',
    data=hoteles_Navarra
).set_title('DISTRIBUCIÓN PLAZAS ESTABLECIM. HOTELEROS EN FUNCIÓN DE SU MODALIDAD Y AGRUPADOS POR CATEGORÍA')
```

Llegamos a las conclusión general de la gráfica anterior, los establecimientos hoteleros con mayor número de plazas son los Hoteles. Pero además ahora podemos decir que dentro de los Hoteles, los que mayor número de plazas tienen son los de 4 estrellas, seguidos a bastante distancia por los de 3 y 5 estrellas.

Si nuestro objetivo es diferenciar la distribución de plazas por categoría dentro de cada modalidad, la gráfica anterior puede parecer algo confusa debido a que los diagramas de caja se ven algo pequeños porque la escala del eje **Y** de plazas es la misma para todas las modalidades, y a que la leyenda con las categorías es demasiado extensa (aparecen todas las categorías y las modalidades de establecimientos que sólo tienen algunas de ellas), lo que hace además que haya demasiados colores que se pueden confundir.  
Para solucionarlo podemos generar un **gráfico con subgráficos**, cada uno de los cuales representa una modalidad:

```python
# Vamos a usar un gráfico con subgráficos de cajas, con varias cajas en cada subgráfico (una por cada categoría 
# de establecimiento hotelero), las cuales nos mostrarán la distribución de la variable plazas para cada categoría 
# dentro de cada modalidad (variable categórica nominal MODALIDAD, variable categórica ordinal CATEGORIA, variable 
# cuantitativa discreta PLAZAS).
```

```python
fig, axarr = plt.subplots(2, 3, figsize=(20, 16))

sns.boxplot(
    x='CATEGORIA',
    y='PLAZAS',
    data=hoteles_Navarra[hoteles_Navarra['MODALIDAD'] == 'Hotel'],
    ax=axarr[0][0]
)
axarr[0][0].set_title("MODALIDAD Hotel")

sns.boxplot(
    x='CATEGORIA',
    y='PLAZAS',
    data=hoteles_Navarra[hoteles_Navarra['MODALIDAD'] == 'Hostal'],
    ax=axarr[0][1]
)
axarr[0][1].set_title("MODALIDAD Hostal")

sns.boxplot(
    x='CATEGORIA',
    y='PLAZAS',
    data=hoteles_Navarra[hoteles_Navarra['MODALIDAD'] == 'Pensión'],
    ax=axarr[0][2]
)
axarr[0][2].set_title("MODALIDAD Pensión")

sns.boxplot(
    x='CATEGORIA',
    y='PLAZAS',
    data=hoteles_Navarra[hoteles_Navarra['MODALIDAD'] == 'Hotel rural'],
    ax=axarr[1][0]
)
axarr[1][0].set_title("MODALIDAD Hotel rural")

sns.boxplot(
    x='CATEGORIA',
    y='PLAZAS',
    data=hoteles_Navarra[hoteles_Navarra['MODALIDAD'] == 'Hotel-apartamento'],
    ax=axarr[1][1]
)
axarr[1][1].set_title("MODALIDAD Hotel-apartamento")
```

Podemos distinguir ahora como en los Hoteles como hemos dicho los que mayor número de plazas tienen son los de 4 estrellas, seguidos por los de 3 y 5 estrellas.  
En los Hostales si son rurales suelen tener menor número de plazas.  
Las Pensiones de dos estrellas suelen tener más plazas que las de una.  
Lo mismo para los Hoteles rurales, al aumentar el número de estrellas de 1 a 4 va aumentando el número de plazas.  
Por último el caso de los Hoteles-apartamento, de los cuales sólo hay dos en este conjunto de datos, en concreto el de 3 estrellas tiene poco más del doble de plazas que el de 4 estrellas.

_**2. Vamos a comprobar la hipótesis de que los establecimientos con mayor número de plazas se encuentran en la zona de la Cuenca de Pamplona, y en concreto en Pamplona capital. Además queremos comprobar si el valor medio del número de plazas de los establecimientos de Pamplona es mayor que para el resto de localidades de la Cuenca de Pamplona.**_

Primero vamos a ver los diferentes valores que puede tomar la variable ZONA:

```python
hoteles_Navarra['ZONA'].unique()
```

Vemos que hay algún registro que tiene el valor NaN para la variable zona. Eliminaremos ese/esos registros para que no de error al representar el gráfico de dispersión siguiente en el que se representan los establecimientos en función de la zona y el número de plazas:

```python
# Vamos a usar un gráfico de dispersión en el que se representarán en el eje X las zonas de la Comunidad Foral 
# de Navarra, y en el eje Y el número de plazas de los establecimientos hoteleros (variable categórica nominal 
# ZONA, variable cuantitativa discreta PLAZAS).
```

```python
# Se genera el gráfico inicial y se almacena una referencia al mismo en 'ax'
ax = hoteles_Navarra.dropna(how='any').plot.scatter(
    x='ZONA', 
    y='PLAZAS', 
    figsize=(20, 15), 
    color='mediumvioletred'
)

# Se emplea la referencia para hacer modificaciones en el gráfico
ax.set_title('Relación entre la zona de Navarra y el número de plazas de los establecimientos', fontsize=20)
```

A la vista del gráfico de dispersión anterior se puede comprobar como los establecimientos con mayor número de plazas se encuentran mayoritariamente en la zona de la Cuenca de Pamplona.  
Se puede ver también, aunque no de manera clara debido al 'overplotting', que hay un gran número de estableciminetos con un número de plazas muy bajo en todas las zonas, pero estos establecimientos de momento no nos interesan para comprobar la primera parte de la hipótesis actual.

Vamos a quedarnos con los establecimientos con 50 plazas o más para verlo más claramente:

```python
# Como en el caso anterior, vamos a usar un gráfico de dispersión en el que se representarán en el eje X las zonas 
# de la Comunidad Foral de Navarra, y en el eje Y el número de plazas de los establecimientos hoteleros siempre 
# que tengan 50 o más plazas (variable categórica nominal ZONA, variable cuantitativa discreta PLAZAS).
```

```python
# Se genera el gráfico inicial y se almacena una referencia al mismo en 'ax'
ax = hoteles_Navarra[hoteles_Navarra['PLAZAS'] >= 50].dropna(how='any').plot.scatter(
    x='ZONA', 
    y='PLAZAS', 
    figsize=(20, 15), 
    color='mediumvioletred'
)

# Se emplea la referencia para hacer modificaciones en el gráfico
ax.set_title('Relación entre zona de Navarra y el número de plazas de establecimientos con más de 50', fontsize=20)
```

Se ve de manera más clara aún que la mayoría de hoteles con un número elevado de plazas (más de 50 en este caso) se encuentran en la zona de la Cuenca de Pamplona.

Vamos a centrarnos ahora en la zona de la Cuenca de Pamplona, para comprobar ahora si es en Pamplona capital donde se encuentran la mayoría de estos establecimientos con elevado número de plazas, tanto de manera general como en valor medio. Para ello generaremos un **gráfico con subgráficos**, tres en este caso, donde se representarán el diagrama de dispersión de establecimientos hoteleros en las localidades de la Cuenca de Pamplona en función del número de plazas, el valor de frecuencia absoluto en las 10 localidades de la Cuenca de Pamplona con más establecimientos de al menos 50 plazas, y el valor medio del número de plazas de todos los establecimientos en cada localidad de la Cuenca de Pamplona:

```python
# Vamos a usar un gráfico con tres subgráficos:
# - El primero será un gráfico de dispersión en el que se representarán en el eje X las localidades de la 
# Cuenca de Pamplona y en el eje Y el número de plazas de los establecimientos hoteleros (variable categórica 
# nominal ZONA, variable categórica nominal LOCALIDAD, variable cuantitativa discreta PLAZAS).
# - El segundo será un gráfico de barras en el que en el eje X se representarán las 10 localidades de la Cuenca de 
# Pamplona con mayor número de establecimientos hoteleros con al menos 50 plazas y en el eje Y la frecuencia 
# absoluta de estos establecimientos (variable categórica nominal ZONA, variable categórica nominal LOCALIDAD, 
# variable cuantitativa discreta PLAZAS).
# - El tercero será un gráfico de barras en el que en el eje X se representarán las localidades de la Cuenca de 
# Pamplona y en el eje Y el valor medio del número de plazas de los establecimientos hoteleros de cada una 
# (variable categórica nominal ZONA, variable categórica nominal LOCALIDAD, variable cuantitativa discreta PLAZAS).
```

```python
# En primer lugar generamos un nuevo DataFrame en el que sólo nos quedamos con los establecimientos hoteleros de 
# la zona de la Cueca de Pamplona:
hoteles_Cuenca_Pamplona = hoteles_Navarra[hoteles_Navarra['ZONA'] == 'Cuenca de Pamplona']

# Generamos los subgráficos que buscamos:
fig, axarr = plt.subplots(3, 1, figsize=(20, 20))

hoteles_Cuenca_Pamplona.plot.scatter(
    x='LOCALIDAD', 
    y='PLAZAS',  
    color='mediumvioletred',
    ax=axarr[0]
)
axarr[0].set_title("Relación localidad/nºplazas")
axarr[0].tick_params(axis='x', labelrotation=90)  # Utilizo este método para girar 90º el texto de las etiquetas 
                                                  # del eje X y que no aparezca en horizontal para que sea legible.

hoteles_Cuenca_Pamplona[hoteles_Cuenca_Pamplona['PLAZAS'] >= 50]['LOCALIDAD'].value_counts().head(10).plot.bar(
    ax=axarr[1]
)
axarr[1].set_title("10 localidades con mayor nº de est. hoteleros con más de 50 plazas en zona Cuenca de Pamplona")
axarr[1].set_ylabel("Frecuencia absoluta")

hoteles_Cuenca_Pamplona.groupby(['LOCALIDAD']).mean().plot.bar(
    ax=axarr[2]
)
axarr[2].set_title("Valor medio del nº plazas de los est. hoteleros según localidad de la zona Cuenca de Pamplona")
axarr[2].set_ylabel("Valor medio de plazas")

plt.suptitle("ZONA DE LA CUENCA DE PAMPLONA", fontsize=20)  # Para poner un título general al conjunto de subplots.

fig.tight_layout()  # Utilizo este método para mantener automáticamente el espacio correcto entre los subgráficos.
```

A la vista de los resultados obtenidos para las localidades de la zona de la Cuenca de Pamplona, podemos concluir que SÍ es en Pamplona capital donde se encuentran la mayoría de establecimientos hoteleros con mayor número de plazas (al menos 50 plazas), pero sin embargo NO es Pamplona capital la localidad con el valor medio del número de plazas de los establecimientos hoteleros más elevado, debido a que al igual que allí están los estableciminetos con mayor número de plazas, también hay un muy elevado número de establecimientos con muy pocas plazas, lo cual hace que baje la media de la capital, en detrimento de otras localidades con menor número de establecimientos pequeños, como por ejemplo la localidad de Mutilva en la que solamente hay un establecimiento hotelero pero tiene más de 150 plazas, con lo cual es la localidad con la media más elevada de la Cuenca de Pamplona.

_**3. Queremos comprobar en primer lugar la hipótesis de que en la zona de los Pirineos la mayor parte de los establecimientos pertenecen a la modalidad de Hostal y a la categoría de rural, debido a que es una zona montañosa sin nucleos grandes de población cercanos. ¿Es la zona de los Pirineos la que tiene el mayor número de establecimientos considerados como rurales (Hostales rurales y Hoteles rurales) de todad la Comunidad Foral de Navarra?**_

Queremos ver en primer lugar si los Hostales rurales son el tipo de establecimiento hotelero que predomina en la zona de los Pirineos.  

Para lo cual utilizaremos en primer lugar una representación de gráfico de barras en la que veremos si los Hostales son, independientemente de su categoría, el establecimineto hotelero más común en la zona de los Pirineos:

```python
# Vamos a usar un gráfico de barras en el que la altura de las barras represente la frecuencia absoluta de 
# establecimientos hoteleros en la zona de los Pirineos en función de su modalidad (variable categórica nominal 
# ZONA, variable categórica nominal MODALIDAD).
```

```python
# Se genera el gráfico inicial y se almacena una referencia al mismo en 'ax'
ax = hoteles_Navarra[hoteles_Navarra['ZONA'] == 'Pirineos']['MODALIDAD'].value_counts().plot.bar(
    figsize=(10, 7), 
    color='brown'
)

# Se emplea la referencia para hacer modificaciones en el gráfico
ax.set_title('Frecuencia de los est. hoteleros en zona Pirineos en función de la modalidad', fontsize=20)
ax.set_xlabel('Modalidad de est. hotelero', fontsize=15)
ax.set_ylabel('Frecuencia absoluta', fontsize=15)
```

A la vista de la gráfica anterior se puede concluir que los Hostales sí son el tipo de establecimiento hotelero predominante en la zona de los Pirineos.

Comprobemos ahora si además son Hostales rurales la mayoría de los establecimientos de la zona:

```python
# Usamos ahora un gráfico de barras en el que la altura de las barras represente la frecuencia absoluta de 
# establecimientos hoteleros en la zona de los Pirineos en función de su modalidad  y categoría (variable 
# categórica nominal ZONA, variable categórica nominal MODALIDAD, variable categórica ordinal CATEGORIA).
```

```python
sns.factorplot(x='MODALIDAD', hue='CATEGORIA', data=hoteles_Navarra[hoteles_Navarra['ZONA'] == 'Pirineos'], 
               kind="count", height=10, aspect=20/10)
```

Podemos comprobar visualizando la gráfica anterior que la primera parte de esta tercera hipótesis planteada se cumple, ya que efectivamente los Hostales con la categoría de rural son el establecimiento hotelero predominante en la zona de los Pirineos.

Vamos a generar ahora un **gráfico con subgráficos** para comprobar si la zona de los Pirineos es la que tiene el mayor número de establecimientos considerados como rurales (Hostales rurales y Hoteles rurales) de toda la Comunidad Foral de Navarra:

```python
# Utilizamos un gráfico de barras para cada subgráfico, en los cuales la altura de las barras represente la 
# frecuencia absoluta de establecimientos hoteleros considerados como rurales en función de la zona a la que 
# pertenecen y la categoría del establecimiento:
# - Subgráfico 1: se consideran establecimientos de la modalidad Hostal (variable categórica nominal ZONA, 
# variable categórica nominal MODALIDAD, variable categórica ordinal CATEGORIA).
# - Subgráfico 2: se consideran establecimientos de la modalidad Hotel rural (variable categórica nominal ZONA, 
# variable categórica nominal MODALIDAD, variable categórica ordinal CATEGORIA).
```

```python
sns.factorplot(x='ZONA', hue='CATEGORIA', col='MODALIDAD', 
               data=hoteles_Navarra[hoteles_Navarra['MODALIDAD'].isin(['Hostal', 'Hotel rural'])], 
               kind="count")
```

Como se puede comprobar viendo las dos gráficas que se han generado, efectivamente la zona de los Pirineos es la que tiene el mayor número de establecimientos considerados como rurales de toda la Comunidad Foral de Navarra, tanto Hostales rurales donde tiene prácticamente el doble que el resto de zonas juntas, como Hoteles rurales, donde destaca sobre todo en los de 2 estrellas, no teniendo ninguno sin embargo de 4 estrellas.  
Esto puede ser debido, como ya se indicaba cuando se planteaba la hipótesis de trabajo, a que se trata de una zona muy montañosa sin núcleos grandes de población cercanos.
